```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\ayael\Downloads\archive (3)\population_by_age_group.csv")

age_groups = ['65+', '25-64 years', '15-24 years', '5-14 years', '0-4 years']

age_group_totals = [data[age_group].sum() for age_group in age_groups]

plt.figure(figsize=(10, 6))
bars = plt.bar(age_groups, age_group_totals, color='skyblue', edgecolor='black')

plt.title('Total Population by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Population')

plt.ticklabel_format(style='plain', axis='y')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{int(yval):,}', ha='center', va='bottom')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
